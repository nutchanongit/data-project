```r
##one variable continuous
c<-ggplot(mtcars, aes(mpg))
c1 <- c + geom_area(stat = "bin")
c2 <- c + geom_density() + geom_dotplot() + geom_histogram(color="red",fill="pink")
c2
```

```r
##one variable discreate/catagorical
mtcars %>%
    select(hp,wt,am) %>%
    mutate(am = ifelse(am==0,"auto","Manual"))%>%
    ggplot(aes(am))+
    geom_bar()
## the graph show count of each catagorical
## we can count like this before in last sprint
mtcars%>%
  select(am)%>%
  mutate(am = ifelse(am==0,"auto","Manual")) %>%
  count(am)
```

```r
##two variable both continuous
base <- ggplot(data = mtcars, mapping = aes(hp,mpg))
base +
  #geom_label(aes(label=hp))
  geom_point(aes(hp,mpg))+
  geom_smooth(method = lm)+
  geom_rug()   ## rug use to show frequency of data at value Y

```

```r
###setting vs mapping
##setting is manual set graphic of graph
##maping is set graphic in aes()

#this is  setting
base + geom_point(col = "red",size= 5,alpha = 0.5)


#this is mapping
base + geom_point(mapping = aes(col=factor(am)))     #แปลงเป็น factor เพื่อให้เป็นข้อมูล discreate
base + geom_point(mapping = aes(col=am))
```

```r
###explore data with chart
set.seed(42)
small_df <- diamonds %>% sample_frac(0.1)
#setting
ggplot(small_df , aes(x = carat , y= price)) +
  geom_point(alpha = 0.3 , col= "#E39786")+
  theme_minimal()
#mapping + setting
ggplot(small_df , aes(x = carat , y= price, col = cut)) +
  geom_point(alpha = 0.3 , size = 3) +
  theme_minimal()

#if you want to show some grade of cut just FILTER
small_df <- diamonds %>% sample_frac(0.1) %>% filter(cut %in% c("Fair", "Premium", "Ideal"))
base2 <-ggplot(small_df , aes(x = carat , y= price, col = cut)) +
  geom_point(alpha = 0.3 , size = 3) +
  theme_minimal()
```
